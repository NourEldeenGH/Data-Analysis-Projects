```python
# import padanas and read the csv file
import pandas as pd
df = pd.read_csv("Amazon Sale Report.csv")

# set a copy for the original file by the name of updated_df
updated_df = df.copy()

# drop n/a cloums
updated_df = updated_df.drop(columns=["fulfilled-by","Unnamed: 22"])

# handling missing data
# removing the duplicates
updated_df = updated_df.drop_duplicates()

# dropping n/a
updated_df = updated_df.dropna()

# # in the status cloumons, correcting the terms:

# First: checking all the possible status by printing them.
#  print(updated_df["Status"].unique())  

# Second: trurning all the results inro lower cases so it can be easily updated without any chance of missing the target.
# updated_df["Status"] = updated_df["Status"].str.strip().str.lower()
# print(updated_df["Status"].unique())  

# Thrid: Chechikng the each uniqe status counter.
# print(updated_df["Status"].value_counts())

# Fourth "Final step here": making a function to set the new terms.
def status_terms(status):
    status = status.lower().strip()  # Ensure it's clean
    
    # Returned statuses
    if "returned to seller" in status or "rejected by buyer" in status or "returning to seller" in status:
        return "Returned"
    
    # Delivered statuses
    elif "delivered to buyer" in status or "picked up" in status:
        return "Delivered"
    
    # Cancelled
    elif "cancelled" in status:
        return "Cancelled"
    
    # Pending
    elif "pending" in status:
        return "Pending"
    
    # Special cases (Lost/Damaged)
    elif "lost in transit" in status or "damaged" in status:
        return "Lost/Damaged"

    # Default to Shipped
    else:
        return "Shipped"

# Apply the function
updated_df["Status"] = updated_df["Status"].apply(status_terms)

# Check if it worked
# print(updated_df["Status"].value_counts())



# swap the colmuns 'ship-postal-code', 'ship-country' as is
# First: Remove the column
col = updated_df.pop("ship-postal-code")
# Second: Insert it at the new position
updated_df.insert(19, "ship-postal-code", col)  
# changing the "ship-postal-code" from (float) into (int)
updated_df["ship-postal-code"] = updated_df["ship-postal-code"].astype(int)

# saving the file to a new updated_data
updated_df.to_csv("cleaned_amazon_sales.csv", index=False)

# updated_df
```
