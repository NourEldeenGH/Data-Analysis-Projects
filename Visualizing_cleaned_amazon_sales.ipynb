# Visualizing the cleaned data by 3 bar charts, 2 pie charts, 1 line chart

### 3 bar charts : represent the (Top 10 Best-Selling Product Categories on Amazon), (Top 10 Cities with the Most Sales), and (Top 10 States with the Most Sales)

### 2 Pie charts : represent the (Sales Distribution by Top 5 States), and (Sales Distribution by Top 5 Cities)

### 1 Line chart : represent the (Daily sales trend on Amazon)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("cleaned_amazon_sales.csv")

category_counts = df["Category"].value_counts().head(10)  # Top 10 categories

# creat the bar chart
plt.figure(figsize=(10,5))
ax = sns.barplot(x=category_counts.index, y=category_counts.values, palette="viridis")

# adding labels over the bars
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}',  
                (p.get_x() + p.get_width() / 2, p.get_height() + 5),  
                ha='center', fontsize=10, color='black')
# Labels & title
plt.xlabel("Product Category")
plt.ylabel("Number of Sales")
plt.title("Top 10 Best-Selling Product Categories on Amazon")

# Show the plot
plt.xticks(rotation=45)
plt.show()
```

```python
# Creating bar chart between ship city and top sales

city_counts = df["ship-city"].value_counts().head(10)

# creat the bar chart
plt.figure(figsize=(10,5))
ax = sns.barplot(x=city_counts.index, y=city_counts.values, palette="viridis")

# adding labels over the bars
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}',  
                (p.get_x() + p.get_width() / 2, p.get_height() + 5),  
                ha='center', fontsize=10, color='black')
# Labels & title
plt.xlabel("City")
plt.ylabel("Number of Sales")
plt.title("Top 10 Cities with the Most Sales")

# Show the plot
plt.xticks(rotation=45)
plt.show()
```

```python
# Creating bar chart between the stats and top sales

state_counts = df["ship-state"].value_counts().head(10)

# creat the bar chart
plt.figure(figsize=(10,5))
ax = sns.barplot(x=state_counts.index, y=state_counts.values, palette="viridis")

# adding labels over the bars
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}',  
                (p.get_x() + p.get_width() / 2, p.get_height() + 5),  
                ha='center', fontsize=10, color='black')
# Labels & title
plt.xlabel("Sate")
plt.ylabel("Number of Sales")
plt.title("Top 10 States with the Most Sales")

# Show the plot
plt.xticks(rotation=45)
plt.show()
```

```python
# Creatin a line chart throw the number of sales and dates

# Step 1: Convert "Date" column to datetime format
df["Date"] = pd.to_datetime(df["Date"], format="%m-%d-%y")

# Step 2: Group by date & count the number of sales
daily_sales = df.groupby("Date")["Order ID"].count()

# Step 3: Plot the sales trend over time
plt.figure(figsize=(12,5))
sns.lineplot(x=daily_sales.index, y=daily_sales.values, marker="o", color="blue")

# Labels & title
plt.xlabel("Date")
plt.ylabel("Number of Sales")
plt.title("Daily sales trend on Amazon")

# Show the plot
plt.xticks(rotation=45)
plt.show()
```

```python
# Creating a Pie Chart for Sales Distribution, by states

state_counts = df["ship-state"].value_counts().head(10)

# Create pie chart
plt.figure(figsize=(8, 8))
plt.pie(state_counts, labels=state_counts.index, autopct="%1.1f%%", colors=sns.color_palette("pastel"))

# Title
plt.title("Sales Distribution by Top 5 States")

# Show the chart
plt.show()
```

```python
# Creating a Pie Chart for Sales Distribution, by cities

city_counts = df["ship-city"].value_counts().head(10)

# Create pie chart
plt.figure(figsize=(8, 8))
plt.pie(state_counts, labels=city_counts.index, autopct="%1.1f%%", colors=sns.color_palette("pastel"))

# Title
plt.title("Sales Distribution by Top 5 Cities")

# Show the chart
plt.show()
```
